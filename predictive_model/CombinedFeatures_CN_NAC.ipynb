### Code below take proportion of each class in train_y dataset to perform label encoding for output

```python
# get a list of output y labels
y_label = train_y.sum(axis=0).index.tolist()

# compute the proportion of each output based on the entire dataset
y_prop = train_y.sum(axis=0) / train_y.shape[0]

# form a dictionary to represent y_label and its proportion
y_dict = {y_label[i]: y_prop[i] for i in range(len(y_label))}

# sort the dictionary
y_dict_sorted = sorted(y_dict.items(), key=lambda item: item[1], reverse=True)

# create a mapping for y_label based on the descending order of y_proportion
y_map = {}

for idx, item in enumerate(y_dict_sorted):
    y_map[item[0]] = idx + 1
    
y_map
```

```python
train_y.head()
```

```python
for k, v in y_map.items():
    train_y[k] = train_y[k].apply(lambda x : v if x >= 1 else 0)
```

```python
for k, v in y_map.items():
    test_y[k] = test_y[k].apply(lambda x: v if x >=1 else 0)
```
